# .index
This attribute provides the index of the dataframe

```python
data.index
```

# .columns

```python
data.columns
```

# .dtypes
It shows the data-type of each columns

```python
data.dtypes
```

# .unique()
In a column, It shows all the unique value. It can be applied on a single column only, not on the whole dataframe.

```python
data['Weather'].unique()
```

# .nunique()
It shows the total no. of unique values in each column. It can be applied on a single column as well as on whole dataframe

```python
data.nunique()
```

# .count
It shows the total no of non-null in each. It can be applied on a single column as well as on whole dataframe.

```python
data.count()
```

# value_counts
In a column, It shows all the unique value with their count. It can be applied on single column only.

```python
data['Weather'].value_counts()
```

# .info()
Provides basic information about the dataframe

```python
data.info()
```
